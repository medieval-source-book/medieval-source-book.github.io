```python
import os

from bs4 import BeautifulSoup

def clean_html(file_path):
    with open(file_path, 'r') as file:
        html = file.read()

    soup = BeautifulSoup(html, 'html.parser')

    # Remove all inline style and class information
    for tag in soup.find_all(True):
        if 'style' in tag.attrs:
            del tag['style']
        if 'span' in tag.attrs:
            del tag['span']

    output_path = file_path.replace('.html', '-clean.html')
    with open(output_path, 'w') as file:
        file.write(str(soup))

    print(f"Cleaned HTML saved to {output_path}")

# Usage
for root, _, files in os.walk('.'):
    for file in files:
        if file.endswith('.html'):
            clean_html(os.path.join(root, file))
```

```python
import os
from bs4 import BeautifulSoup

def clean_html(file_path):
    with open(file_path, 'r') as file:
        html = file.read()

    soup = BeautifulSoup(html, 'html.parser')

    # Remove all inline style and class information
    for tag in soup.find_all(True):
        if 'style' in tag.attrs:
            del tag['style']
        if 'span' in tag.attrs:
            del tag['span']

    # Insert a line break after each tag that starts with </
    for tag in soup.find_all(True):
        if tag.name.startswith('</'):
            tag.insert_after('\n')

    output_path = file_path.replace('.html', '-clean.html')
    with open(output_path, 'w') as file:
        file.write(str(soup))

    print(f"Cleaned HTML saved to {output_path}")

# Usage
for root, _, files in os.walk('.'):
    for file in files:
        if file.endswith('.html'):
            clean_html(os.path.join(root, file))
```
